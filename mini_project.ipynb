```python
import pandas as pd
```

```python
df = pd.read_csv('/student_performance.csv')
```

```python
print("-- DATASET OVERVIEW --")
print("Total Students:", df.shape[0])
print("Total Columns:", df.shape[1])
print("Total Departments:", df['department'].nunique())
```

```python
print("------- DEPARTMENT WISE ANALYSIS -----")
print(df['department'].value_counts())
```

```python
print("---- SCORE ANALYSIS ----")
print("Highest Math Score:" , df['math_score'].max())
print("Lowest Math Score:" , df['math_score'].min())
print("Average Math Score:" , df['math_score'].mean())
print("Average Programming Score:" , df['programming_score'].mean())
print("Average Attendance Percentage:" , df['attendance_percentage'].mean())
```

```python
df['total_score'] = (
    df['math_score'] +
    df['science_score'] +
    df['english_score'] +
    df['programming_score']
)

print("---- GENDER WISE ANALYSIS ----")

gender_score = df.groupby('gender')['total_score'].mean().round(2)

print(gender_score)
```

```python
print("----- TOP PERFORMERS ----")
top5 = df.sort_values('total_score',ascending = False).head(5)
for index , row in top5.iterrows():
    print(f"Name: {row['name']:<20}, Department: {row['department']:<18}, Total Score: {row['total_score']}")
```

```python
print("---- STUDENTS WHO NEED ATTENTION ----")

low_att = df[df['attendance_percentage'] < 75][
    ['name', 'department', 'attendance_percentage']
]

if len(low_att) > 0:
    print(low_att.to_string(index=False))
else:
    print("All Students have attendance above 75")
```
