```python
print("---- HANDLE MISSING VALUES ----")
print("Before Fixing Missing Values:")
print(df.isnull().sum())

print("\nTotal Missing Values:", df.isnull().sum().sum())

# Fill numeric column with mean
df['quantity'].fillna(df['quantity'].mean(), inplace=True)

# Fill categorical columns with mode
df['customer_name'].fillna(df['customer_name'].mode()[0], inplace=True)

df['product'].fillna(df['product'].mode()[0], inplace=True)

df['category'].fillna(df['category'].mode()[0], inplace=True)

print("\nAfter Fixing Missing Values:")
print(df.isnull().sum())

print("\nTotal Missing Values:", df.isnull().sum().sum())
```

```python
print("Before Duplication:", len(df))
df.drop_duplicates(inplace=True)
print("After Duplication:", df.duplicated().sum())
```

```python
# Convert Order Date column to datetime
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst=True)

# Extract year, month number, and month name
df['Year'] = df['order_date'].dt.year
df['Month'] = df['order_date'].dt.month
df['Month_Name'] = df['order_date'].dt.month_name()

# Show output
print(df[['order_date', 'Year', 'Month', 'Month_Name']].head())
```

```python
df['customer_name'] = (
    df['customer_name']
    .str.strip()
    .str.title()
)
mask = (df['product'] == 'keyboard') & (df['category'] == 'Electronics')
df.loc[mask, 'category'] = 'Accesories'
print("After fix: Unique Characters" , df['category'].unique())
```

```python
# Convert quantity and unit_price to integer

df['quantity'] = df['quantity'].astype(int)

df['unit_price'] = df['unit_price'].astype(int)

# Create revenue column

df['revenue'] = (
    df['quantity'] * df['unit_price']
    ).astype(int)

# Display revenue

print(
    df[
        ['customer_name', 'product', 'quantity', 'unit_price', 'revenue']
    ].head()
)

print("Total Revenue:", df['revenue'].sum())
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        # If the column is entirely NaN, mode() will be empty.
        # In this case, fill with a placeholder like 'Unknown'.
        if df[col].isnull().all():
            df[col].fillna('Unknown', inplace=True)
        # Otherwise, if there are non-NaN values, use the mode.
        elif not df[col].mode().empty:
            df[col].fillna(df[col].mode()[0], inplace=True)
    else: # Numeric columns
        # If the column is entirely NaN, df[col].mean() will be NaN.
        # In this case, fill with 0 (or another appropriate default).
        if df[col].isnull().all():
            df[col].fillna(0, inplace=True)
        # Otherwise, if there are non-NaN values, use the mean.
        else:
            df[col].fillna(df[col].mean(), inplace=True)
```
