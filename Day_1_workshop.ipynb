```python
df['mean'] = (
    df['math_score']+
    df['science_score']+
    df['english_score']+
    df['programming_score']
)/4

df[(df['mean']>90) & (df['attendance_percentage']>90)]
```

```python
df['total_score'] = (
    df['math_score'] +
    df['science_score'] +
    df['english_score'] +
    df['programming_score']
)

df.groupby('department')['total_score'].agg(
    ['mean']
).sort_values(by = 'mean', ascending = False).reset_index()
```
